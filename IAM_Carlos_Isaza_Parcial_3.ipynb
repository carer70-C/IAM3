### (a) 

Verifique el límite del teorema 2.4 en los tres casos del ejemplo 2.2.

### Solución

El teorema 2.4 dice que dado un espacio de hipótesis $H$, si $k\in\mathbb{N}$ es un punto de de quiebre para $H$ entonces la función de crecimiento $m_{H}(N)$ tiene como cota superior a $\sum^{k-1}_{i=0}{N \choose i}$ (esto para cualquier entero positivo $N$), es decir que, $m_{H}(N) \le \sum^{k-1}_{i=0}{N \choose i}$. 

#### (i) Rayos positivos:

$H$ consta de todas las hipótesis en una dimensión de la forma $h(x)=signo(x-a)$, con $a\in\mathbb{R}$ fijo.
Como se discutió en el ejemplo 2.2  $m_{H}(N)=N+1$. Veamos el punto de quiebre $k$,

rayos.PNG

 
con $N=1$, si $a$ esta en la zona $1$ entonces $h(x_1)=+1$, si esta en la zona $2$, $h(x_1)=-1$; pero con $N=2$, si $a$ esta en la zona $1$ entonces $h(x_1)=+1$ y $h(x_2)=+1$, si esta en la zona $2$ entonces $h(x_1)=-1$ y $h(x_2)=+1$ , si esta en la zona $3$ entonces $h(x_1)=-1$ y $h(x_2)=-1$; es decir que no existe $h\in H$ tal que $h(x_1)=+1$ y $h(x_2)=-1$, luego k=2.

Verificando la cota, tenemos que  $m_{H}(N)=N+1=\frac{N(N-1)!}{1!(N-1)!}+\frac{N!}{N!}=\frac{N!}{1!(N-1)!}+\frac{N!}{0!(N-0)!}={N \choose 1}+{N \choose 0}=\sum^{2-1}_{i=0}{N \choose i}$.

#### (ii) Intervalos positivos: 

$H$ consta de todas las hipótesis en una dimensión que son positivos dentro de algún intervalo y negativos en otros lugares. En el ejemplo 2.2 vimos que $m_{H}(N)={N+1 \choose 2}+1$. Veamos el punto de quiebre $k$,

intervalos0.PNG
 
En el anterior esquema se observa que $H$ particiona a cualquier conjunto de puntos de tamaño $2$, por lo que $k=2$ no es un punto de quiebre (en realidad para que $k$ no sea punto de quiebre solo se necesita que $H$ particione a un conjunto de tamaño $k$, pero, en este caso, por el tipo de espacio unideimensional con orden total, si $H$ particiona un conjunto de tamaño $k$ tambien lo hace para cualquier conjunto del mismo tamaño).

intervalos.PNG
 
En el esquema de arriba se observa que no existe $h\in H$ tal que $h(x_1)=+1$, $h(x_2)=-1$ y $h(x_3)=+1$, por tanto $k=3$ es un punto de quiebre.
 
 Verificando, tenemos que $m_{H}(N)={N+1 \choose 2}+1=\frac{N(N+1)}{2} + 1= \frac{N(N-1)}{2} + N+1=\sum^{3-1}_{i=0}{N \choose i}$.
 
 #### (iii) Conjuntos convexos: 
 
$H$ consta de todas las hipótesis en dos dimensiones que son positivas dentro de algún conjunto convexo y negativas en otras partes. En el ejemplo 2.2 vimos que  $m_{H}(N)=2^{N}$, por tanto no existe punto de quiebre $k\leq N$; sin embargo observe que con $k>N$, $k$ cumple vaciamente la definición 2.3 de ser punto de quiebre de $H$, ya que no existe un subconjunto de tamaño $k>N$ en un conjunto de tamaño $N$, además note que ${N \choose i}=0$ para $i>N$, luego tenemos que $\sum^{\infty -1}_{i=0}{N \choose i}=\sum^{N+1-1}_{i=0}{N \choose i}$, por el binomio de Newton tenemos que,  $\sum^{N}_{i=0}{N \choose i}=\sum^{N}_{i=0}{N \choose i}1^{N-i}1^{i}=(1+1)^N=2^N=m_{H}(N)$, verificando la cota del teorema 2.4.
 
### (b) 

¿Existe un conjunto de hipótesis para el cual $m_H(N)=N+2^{[\frac{N}{2}]}$ (donde $[\frac{N}{2}]$ es el entero más grande $\leq \frac{N}{2}$)?

### Solución

Note que como $m_H(N)=N+2^{[\frac{N}{2}]}$ entonces $m_H(N)<2^{N}$ para $N\geq3$, por tanto debería existir un punto de quiebre $k\leq N$ (dado un subconjunto de tamaño $k$, $H$ no lo puede particionar; si no existiera tal $k$, $m_H(N)=2^{N}$). Luego, por el teorema 2.4, $m_H(N)=N+2^{[\frac{N}{2}]}\leq \sum^{k-1}_{i=0}{N \choose i}$; pero esto es una contradiccón porque el lado izquierdo de la anterior inecuación crece de forma exponencial y el lado derecho de forma polinomial, es decir que para cualquier $k$ existe un $N$ lo suficientemente grande tal que   $m_H(N)=N+2^{[\frac{N}{2}]}> \sum^{k-1}_{i=0}{N \choose i}$, de lo que se concluye que no existe tal $H$. 

Acontinución se grafican algunos ejemplos.

```python
for k in range(1,7):
    N=k
    j=0
    MhVector=[]
    CotaMhVector=[]
    Mh=N+2**(math.modf(N/2)[1])
    CotaMh=sum([math.factorial(N)/(math.factorial(i)*math.factorial(N-i)) for i in range(0,k)])
    #Este ciclo calcula el N para que la función de crecimiento cruce a la cota del teorema 2.4
    while Mh<=CotaMh or j<6:
        MhVector.append(Mh)
        CotaMhVector.append(CotaMh)
        N+=1
        if Mh>CotaMh:
            j+=1   
        Mh=N+2**(math.modf(N/2)[1])
        CotaMh=sum([math.factorial(N)/(math.factorial(i)*math.factorial(N-i)) for i in range(0,k)])
    x1=list(range(k,N))
    plt.plot(x1,MhVector, 'r',label="$m_{H}(N)$")
    plt.plot(x1,CotaMhVector, 'b',label="$Cota$")
    plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
    plt.grid()
    plt.title(f"k={k}", 
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 18})
    plt.xlabel('$N$')
    plt.show()
```

Note que si existiera un punto de quiebre $k\leq N$, el candidato más pequeño sería $k=3$, que como se ve en la grafica correspondiente, es donde el teorema 2.4 se cumple para $N\leq 13$; pero a partir de $N>13$ se obtine $m_H(N)=N+2^{[\frac{N}{2}]}> \sum^{3-1}_{i=0}{N \choose i}$, confirmando nuestro análisis anterior en que concluimos que tal $H$ no existe.

### Ejercicio 2.5

Supongamos que tenemos un modelo de aprendizaje simple cuya función de crecimiento es $m_H(N)=N+1$, por lo tanto, $d_{vc}=1$. Use la cota VC (2.12) para estimar la probabilidad de que $E_{out}$ esté $0.1$ a la derecha de $E_{in}$ con un entrenamiento de $100$ ejemplos. (Sugerencia: la estimación será ridícula).

### Solución

La cota VC nos dice que con una probabilidad de al menos de $1-\delta$,

$$E_{out}\leq E_{in}+\sqrt{\frac{8}{N}\ln{\frac{4m_{H}(2N)}{\delta}}}$$

Lo que nos pide el ejercicio es que estimemos la probabilidad de que $\sqrt{\frac{8}{N}\ln{\frac{4m_{H}(2N)}{\delta}}}=s$ con $s=0.1$, $N=100$ y $m_H(N)=N+1$. Sabemos que la probailidad es mayor o igual a  $1-\delta$, por tanto nuestra tarea es encontrar el valor de $\delta$.

Despejando $\delta$ de la última ecuación, obtenemos que:

$$\frac{Ns^2}{8}=\ln{\frac{4m_{H}(2N)}{\delta}}$$

$$e^{\frac{Ns^2}{8}}=\frac{4m_{H}(2N)}{\delta}$$

$$\delta=\frac{4m_{H}(2N)}{e^{\frac{Ns^2}{8}}}$$

Ahora calculemos $\delta$:       

```python
s=0.1
N=100
Mh=2*N+1
numerador=4*Mh
denominador=np.exp(N*s**2/8)
delta=numerador/denominador
print(f'𝛿={delta}, por tanto la probabilidad buscada es mayor o igual a {1-delta}.')
```

Esto nos indica que la probabilidad de que $E_{out}$ esté $0.1$ a la derecha de $E_{in}$ es mayor o igual a cero, lo cual ya lo sabíamos de antemano. 

Este ejercicio nos ilustra que la cota VC es bastante holgada.

Tenemos acceso a dos estimaciones: $E_{in}(g)$, el error en la muestra en los 400 ejemplos de entrenamiento; y, $E_{test}(g)$, el error de prueba en los 200 ejemplos de prueba que se dejaron de lado.

### (a) 

Usando una tolerancia de error del 5% ($\delta=0.05$), ¿qué estimación tiene la 'barra de error' más alta?

### Solución

La variente de la desiguldad de Hoeffding para un conjunto de hipótesis finito es,

$$Pr\left[|v-\mu|>\epsilon\right]\leq2Me^{-2N\epsilon^{2}}.$$

Si $\delta$ es la tolerancia de la "mala probabilidad", despejando $\epsilon$ obtenemos,

$$\epsilon=\sqrt{\frac{1}{2N}\ln{\frac{2M}{\delta}}}.$$

Por otra parte, tenemos que con una probabilidad menor o igual a $\delta$,

$$|v-\mu|>\epsilon$$

$$-\epsilon>v-\mu>\epsilon$$

entonces, 

$$-v-\epsilon>-\mu.$$

Luego, con una probabilidad mayor o igual $1-\delta$,

$$\mu \leq v+\epsilon.$$

Ahora, si $g\in H$ es la hipótesis escogida por el algoritmo de aprendizaje, con $v\in\{E_{in}(g),E_{test}(g)\}$, $\mu=E_{out}(g)$ y $\epsilon=\sqrt{\frac{1}{2N}\ln{\frac{2M}{\delta}}}$ la anterior desigualdad queda,

$$E_{out}(g) \leq v+\sqrt{\frac{1}{2N}\ln{\frac{2M}{\delta}}}.$$

Si $v=E_{in}(g)$ entonces, como el algoritmo de aprendizajese usó 400 datos de entrenamiento y un conjunto de 1000 hipótesis, $N=400$ y $M=1000$. Luego, con $\delta=0.05$, tenemos que: 

```python
N=400
M=1000
delta=0.05
def epsilon(M,N,delta):
    return np.sqrt(1/(2*N)*np.log(2*M/delta))
ein=epsilon(M,N,delta)
print(f'Con una probabilidad mayor o igual a {1-delta}, 𝐸𝑜𝑢𝑡(𝑔)≤𝐸𝑖𝑛(𝑔)+{ein}.')
```

Si $v=E_{test}(g)$ entonces, como los N=200 datos con que se midió $E_{test}(g)$ no fueron usados por el algoritmo de aprendizaje, $M$ corresponde solo al conjunto unitario $\{g\}$, es decir que $M=1$. Luego, con $\delta=0.05$, tenemos que:

```python
N=200
M=1
delta=0.05
etest=epsilon(M,N,delta)
print(f'Con una probabilidad mayor o igual a {1-delta}, 𝐸𝑜𝑢𝑡(𝑔)≤𝐸𝑡𝑒𝑠𝑡(𝑔)+{etest}.')
```

Por tanto, $E_{test}(g)$ esta más cerca de $E_{out}(g)$. Este ejercicio ilustra que para predicir cómo se va a comportar $g$ fuera de la muestra de entrenamiento, es mejor usar $E_{test}(g)$ que $E_{in}(g)$.

### Problema 2.1

En la ecuación (2.1), establezca $\delta=0.03$ y sea

$$\omega(M,N,\delta)=\sqrt{\frac{1}{2N}\ln{\frac{2M}{\delta}}}.$$

(a) Para $M = 1$, ¿cuántos ejemplos necesitamos para hacer $\omega\leq 0.05$?

(b) Para $M = 100$, ¿cuántos ejemplos necesitamos para hacer $\omega\leq 0.05$?

(c) Para $M = 10,000$, ¿cuántos ejemplos necesitamos para hacer $\omega\leq 0.05$? 

### Solución

La ecución (2.1) es $$E_{out}(g) \leq E_{in}(g)+\sqrt{\frac{1}{2N}\ln{\frac{2M}{\delta}}}.$$

Lo que nos pide el ejercicio ese calcular $N$ tal que $$\sqrt{\frac{1}{2N}\ln{\frac{2M}{\delta}}}\leq \epsilon$$

con $\delta =0.03, \epsilon=0.05$ y $M\in\{1,100,10000\}$. Despejando $N$ de la anterior desicualdad, obtenemos que

$$N\geq \frac{1}{2\epsilon^2}\ln(\frac{2M}{\delta}).$$

Acontinución se utiliza la anterior desigualdad para calcular $N$ en cada uno de los casos:

```python
delta = 0.03
epsi = 0.05
#se define M con más elementos para ver mejor como se comprta el numero de datos necesarios
M=[1,10,100,1000,10000,100000,1000000]

Nvector=[]
def Ndatos(M, delta, epsi):
    return 1/(2*epsi**2)*np.log(2*M/delta)

for i in M:
    N = math.modf(Ndatos(i, delta, epsi))[1]+1
    Nvector.append(N)
    print(f"Para M={i} se necisitan al menos {N} datos.")
print("Las diferencias entre los N´s encontrados son:")
for i in range(0,len(Nvector)-1):
    print(f'{Nvector[i+1]-Nvector[i]}')
#se grafica M vs N
plt.plot(M,Nvector, 'r',marker='o',linewidth=0,label="$m_{H}(N)$")
plt.xlabel('$M$')
plt.ylabel('$N$')
plt.grid()
```

Naturalmente, se observa que a mayor cantidad de hipótesis, $H$ es más complejo y el algoritmo de aprendizaje necesita mayor cantidad de datos para alcanzar el objetivo, $\sqrt{\frac{1}{2N}\ln{\frac{2M}{\delta}}}\leq \epsilon$; si $M$ crece en un factor de $10$ entonces $N$ crece constantemente un valor $461$. 

Por otra parte, este problema nos recuerda que si $M$ es demasiado grande o infinito, y no se tubiera encuenta que varias hipótesis se pueden comportar de la misma forma sobre los datos de entrenamiento, el problema de aprendizaje sería intratable. 

### Problema 2.12 

Para una $H$ con $d_{vc}=10$, ¿qué tamaño $N$ se necesita (según lo prescrito por el límite de generalización) para tener un $95\%$ de confianza de que su error de generalización sea como máximo $0.05$? 

### Solución

El límite de generalización nos dice que con probabilidad mayor o igual a $1-\delta$,

$$E_{out}(g) \leq E_{in}(g)+\sqrt{\frac{8}{N}\ln{\frac{4m_{H}(2N)}{\delta}}}.$$

Lo que indica el problema es que estimemos $N$ teniendo en cuenta que $1-\delta=95\%$ (es decir con $\delta=0.05$), $\sqrt{\frac{8}{N}\ln{\frac{4m_{H}(2N)}{\delta}}}\leq \epsilon=0.05$ y que para $H$, $d_{vc}=10$. 

Para poder usar la condición de que $d_{vc}=10$, recordemos que por el teorema 2.4, $m_{H}(N) \leq \sum^{k-1}_{i=0}{N \choose i}$ con $k$ punto de quiebre de $H$, y, como por definición, $d_{vc}$ es el numero más grande tal que $m_h(d_{vc})=2^{d_{vc}}$, $m_h(N)<2^{d_{vc}+1}$; es decir que, $k=d_{vc}+1$ es el punto de quiebre de $H$ (de hecho el más pequeño). Por tanto $m_{H}(N) \leq \sum^{d_{vc}}_{i=0}{N \choose i}$. Además, $\sum^{d_{vc}}_{i=0}{N \choose i}$ es un polinomio en $N$ de grado $d_{vc}$ y se puede demostrar que $\sum^{d_{vc}}_{i=0}{N \choose i}\leq N^{d_{vc}}+1$, por tanto $m_{H}(N) \leq N^{d_{vc}}+1$. Utilizando este resultado tenemos que,

$$\sqrt{\frac{8}{N}\ln{\frac{4((2N)^{d_{vc}}+1)}{\delta}}}\leq \epsilon.$$

De la cual obtenemos que,

$$N\geq \frac{8}{\epsilon^2}\ln{\frac{4((2N)^{d_{vc}}+1)}{\delta}}.$$

De la anterior desigualdad podemos estimar $N$ dandole un valor inicial, por ejemplo de $N=1000$, e iterar hasta encontrar una convergencia con alguna tolerancia:

```python
delta=0.05
epsi=0.05
dvc=10

N_inicial=1000
N_iterado=0
tolerancia=0.1
control=0
def Ndatos2(delta, epsi,dvc,N):
    return 8/(epsi**2)*np.log(4*((2*N)**dvc+1)/delta)

res=np.abs(N_iterado-N_inicial)
while res>tolerancia and  control<10000:
    N_iterado=Ndatos2(delta, epsi,dvc,N_inicial)
    res=np.abs(N_iterado-N_inicial)
    N_inicial=N_iterado
    control+=1
N_final=math.modf(N_iterado)[1]+1

print(f"Para tener un  95%  de confianza de que el error de generalización sea como máximo  0.05 se necisitan al menos {N_final} datos.")
```

Veamos qué sucede con $N$ a medida que crece $d_{vc}$ con valor constante de $5$:

```python
dvc=[10,15,20,25,30]
N_final_vector=[]
for i in dvc:
    control=0
    N_inicial=1000
    N_iterado=0
    res=np.abs(N_iterado-N_inicial)
    while res>tolerancia and  control<10000:
        N_iterado=Ndatos2(delta, epsi,i,N_inicial)
        res=np.abs(N_iterado-N_inicial)
        N_inicial=N_iterado
        control+=1
    N_final=math.modf(N_iterado)[1]+1
    N_final_vector.append(N_final)
    print(f"Para dvc={i} se necisitan al menos {N_final} datos.")

print("Las diferencias entre los N´s encontrados son:")
for i in range(0,len(N_final_vector)-1):
    print(f'{N_final_vector[i+1]-N_final_vector[i]}')
    
plt.plot(dvc,N_final_vector, 'r',marker='o',linewidth=0,label="$m_{H}(N)$")
plt.xlabel('$dvc$')
plt.ylabel('$N$')
plt.grid()
```

Observamos que $N$ es proporcional a $d_{vc}$; pero a diferencia a lo observado al punto 2.1, donde $M$ necesitaba crecer por un factor de 10 para que $N$ creciera un valor costante, aquí $d_{vc}$ solo necesita crecer una costante de $5$ para que $N$ cresca un valor aproximadamente costante. 

En este punto vemos que si usamos $H$ infinito, la $d_{vc}$ nos permite manipularlo pero la cantidad de datos requeridas es bastante alta comparada con un $H$ finito. Por ejemplo, con lo parametros de este ejercicio, para que se necesiten $452957$ datos en un $H$ finito, $M$ tendría que ser extremadamente grande:

### Problema 2.16 

En este problema, consideraremos $X = \mathbb{R}$. Es decir, $x\in\mathbb{R}$ es una variable unidimensional. Para un conjunto de hipótesis 

$$H=\{h_c:h_c(x)=signo(\sum^{D}_{i=0}{c_ix^i}) ~con~c=\{c_0,...,c_D\}\in \mathbb{R^{D+1}}\},$$

demuestre que la dimensión VC de $H$ es exactamente $D+1$ mostrando (a) y (b).
 
### (a) 

Hay $D+1$ puntos que son particionados por $H$.

### Solución

Dado un conjunto de $D+1$ puntos, $S:=\{x_0,...,x_D\}$, podemos suponer que $x_0< x_1<x_2<...<x_D$, por tanto la  forma general de ubicarlos en la recta real es uno tras otro de menor a mayor (esto nos indica que, en este caso, la distribución de los puntos en el espacio no afecta el hecho de que $H$ rompa o no a $S$, es más, la forma en que se comporta $H$ sobre un $S$ en particular nos dice como se comporta sobre cualquier otro conjunto de puntos de ese mismo tamaño). 

Por otra parte, dado $x_t\in S$, $f(x_t)\in\{+1,-1\}$ con $f$ nuestra función objetivo, para que $h_c(x_t)=f(x_t)$ necesitamos que $\sum^{D}_{i=0}{c_ix_t^i}>0$, si $f(x_t)=+1$, y que $\sum^{D}_{i=0}{c_ix_t^i}<0$, si $f(x_t)=-1$.

Note que  $\sum^{D}_{i=0}{c_ix^i}$ es un polinomio de grado a lo más $D$, entonces nuestra tarea se puede reducir a encontrar los coeficientes $c_i$ tales que el polinomio $\sum^{D}_{i=0}{c_ix^i}$ pase por el conjunto de puntos $\{(x_0,f(x_0)),...,(x_D,f(x_D))\}$; pero este polinomio existe, de hecho uno de los resultados importantes del análisis numerico dice que dados $D+1$ puntos en $\mathbb{R}^2$ existe un polinomio a lo más de grado $D$ que pasa por los $D+1$ puntos, y para encontrar dicho polinomio se usan metodos de interpolación polinomica. 

La demostración de la existencia de ese este polinomio se basa en que el sistema


\begin{equation*}
\begin{bmatrix} 
y_0\\
.\\
.\\
.\\
y_D
\end{bmatrix}
=
\begin{bmatrix} 
x_{0}^0 & x_{0}^1 & ... & x_{0}^D \\
x_{1}^0 & x_{1}^1 & ... & x_{1}^D \\
. & . & ... &. \\
. & . & ... &. \\
. & . & ... &. \\
x_{D}^0 & x_{D}^1 & ... & x_{D}^D
\end{bmatrix}
\begin{bmatrix}
c_0 \\
. \\
. \\
. \\
c_D
\end{bmatrix},
\end{equation*}

donde en nuestro caso $y_t=f(x_t)\in\{+1,-1\}$, tiene solución ya que la matriz 

\begin{bmatrix} 
x_{0}^0 & x_{0}^1 & ... & x_{0}^D \\
x_{1}^0 & x_{1}^1 & ... & x_{1}^D \\
. & . & ... &. \\
. & . & ... &. \\
. & . & ... &. \\
x_{D}^0 & x_{D}^1 & ... & x_{D}^D
\end{bmatrix}

es la matriz de Vandermonde y es invertible siempre que $x_t\neq x_j$ con $t\neq j$ (una pureba de esto se puede ver en https://es.wikipedia.org/wiki/Matriz_de_Vandermonde).

A continuación se ilustra este item eligiendo aleatoriamente dos conjuntos de puntos de tamaño $D+1$; el primer conjunto representa a $S:=\{x_0,...,x_D\}$ y el segundo conjunto, $W:=\{w_0,...,w_D\}$, representa el conjunto $\{f(x_0),...f(x_D)\}$. Aun que el conjunto $W$, generado aleatoriamente,  no es precisamente de $+1$´s y $-1$´s si representa a $\{f(x_0),...f(x_D)\}$ ya que basta con que sus elementos sean mayores o menores a $0$ tales que $signo(w_t)=f(x_t)$. Despues, utilizando metodos de interpolación predeterminados de python,  se encuentra el vector de coeficientes $c$ tales que $\sum^{D}_{i=0}{c_ix_t^i}=w_t$ (para $0\leq t\leq D$), es decir que se encuentra un polinomio que pasa por los puntos $(x_0,w_0),...,(x_D,w_D)$. Y por último se grafican los datos.

```python
#se define una función que tiene por entrada un numero de puntos y como salidae todos los datos necesarios 
#para graficar una interpolación
def interpol(puntos):
    #se generan aleatoriamente la muestra de puntos S con su respectivo conjunto W 
    S=np.random.randint(-100,100, size=(puntos, 1)).flatten()
    W=np.random.randint(-100,100, size=(puntos, 1)).flatten()
    # se calcula los coeficientes del vector c
    C = np.polyfit(S,W,len(S)-1)
    # se generan los datos a evaluar en el polinomio
    X = np.linspace(min(S),max(S))
    #se evalua el polinomio
    pol = np.polyval(C,X)
    #se separan los puntos de S a los que f les asigna un valor positivo de los que les asigna un valor negativo
    S_positivos=[]
    S_negativos=[]
    j=0
    for i in W:
        if i>0:
            S_positivos.append(S[j])
        else:
            S_negativos.append(S[j])
        j+=1
    
    return S,W, X,pol, S_positivos, S_negativos

# se define el valor para D, con valores mayores o iguales a 6 el polinomio resultante
# tiene grandes oscilaciones y se  vuelve muy difícil de graficar
D=5
puntos=D+1

# se grafincan los datos
interpol=interpol(puntos)
S=interpol[0]
W=interpol[1]
X=interpol[2]
pol=interpol[3]
S_positivos=interpol[4]
y_positivos=[0]*len(S_positivos)
S_negativos=interpol[5]
y_negativos=[0]*len(S_negativos)
plt.plot(S,W,marker='o',linewidth=0,c='b',label="$(x_t,y_t)$")
plt.plot(X,pol,c='y',label="$Pol$")
plt.plot(S_positivos,y_positivos,marker='o',linewidth=0,c='g',label="$f(x_t)=+1$")
plt.plot(S_negativos,y_negativos,marker='o',linewidth=0,c='r',label="$f(x_t)=-1$")
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.title("$H$ PARTICIONA $D+1$ PUNTOS", 
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 18})
```

Este item nos dice que $d_{vc}(H)\geq D+1$.

### (b) 

No hay $D+2$ puntos que sean particionados por $H$. 

### Solución

Dado $S=\{x_0,...x_{D+1}\}$ elijamos como dicotomía la que intercala $+1$ y $-1$, esto es, 

$$F:=\{f(x_t): f(x_t)=+1 ~si~t~es~par~y f(x_t)=-1 ~si~t~es~impar, 0\leq t\leq D+1 \}.$$

Por ejemplo si $D$ fuera impar entonces $F=\{+1,-1,...,+1\}$ con $D+2$ elementos.

De este modo, cualquier polinomio que pase por los puntos $(x_0,f(x_0)),..,(x_{D+1},f(x_{D+1}))$ va a tener un cambio de signo entre cualesquiera para de puntos contiguos de $S$, es decir que, $signo(\sum^{D}_{i=0}{c_ix_t^i})\neq signo(\sum^{D}_{i=0}{c_ix_{t+1}^i})$ con $0\leq t\leq D$. Por tanto, dicho polinomio va a tener $D+1$ cambios de signo, lo que indica que necesariamente tendría $D+1$ raices reales, es decir que sería de grado $D+1$ y por tanto no pertenecería a $H$.

Analogo a (a), se ilustra con una grafica, con la diferencia que el vector aleatorio $W$ tiene la condición de que $w_t>0$ si $t$ es par y $w_t<0$ si $t$ es impar.

```python
#se define una función que tiene por entrada un numero de puntos y como salidae todos los datos necesarios 
#para graficar una interpolación
def interpol(puntos):
    #se generan aleatoriamente la muestra de puntos S con su respectivo conjunto W 
    S=np.random.randint(-100,100, size=(puntos, 1)).flatten()
    S.sort()
    W_aux=np.random.randint(-100,100, size=(puntos, 1)).flatten()
    W=[]
    K=0
    for i in W_aux:
        if K%2==0 and i>0:
            W.append(i)
        if K%2==0 and i<0:
            W.append(-i)
        if K%2==1 and i>0:
            W.append(-i)
        if K%2==1 and i<0:
            W.append(i)
        K+=1
    W=np.array(W)
    
    # se calcula los coeficientes del vector c
    C = np.polyfit(S,W,len(S)-1)
    # se generan los datos a evaluar en el polinomio
    X = np.linspace(min(S),max(S))
    #se evalua el polinomio
    pol = np.polyval(C,X)
    #se separan los puntos de S a los que f les asigna un valor positivo de los que les asigna un valor negativo
    S_positivos=[]
    S_negativos=[]
    j=0
    for i in W:
        if i>0:
            S_positivos.append(S[j])
        else:
            S_negativos.append(S[j])
        j+=1
    
    return S,W, X,pol, S_positivos, S_negativos

# se define el valor para D, con valores mayores o iguales a 6 el polinomio resultante
# tiene grandes oscilaciones y se  vuelve muy difícil de graficar
D=5
puntos=D+2

# se grafincan los datos
interpol=interpol(puntos)
S=interpol[0]
W=interpol[1]
X=interpol[2]
pol=interpol[3]
S_positivos=interpol[4]
y_positivos=[0]*len(S_positivos)
S_negativos=interpol[5]
y_negativos=[0]*len(S_negativos)
plt.plot(S,W,marker='o',linewidth=0,c='b',label="$(x_t,y_t)$")
plt.plot(X,pol,c='y',label="$Pol$")
plt.plot(S_positivos,y_positivos,marker='o',linewidth=0,c='g',label="$f(x_t)=+1$")
plt.plot(S_negativos,y_negativos,marker='o',linewidth=0,c='r',label="$f(x_t)=-1$")
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.title("$H$ NO PARTICIONA $D+2$ PUNTOS", 
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 18})
```

La anterior grafica nos ilustra que el polinomio encontrado siempre intersecta al eje $x$ en $D+1$ puntos (en este caso $D=5$), lo que indica que es un polinomio de al menos grado $D+1$, es decir que el vector de coeficientes encontrado $\bar{c}\in\mathbb{R^{D+1}}$,por tanto $h_{\bar{c}} \notin H$, luego $H$ no particiona a $S$.

Este item nos muestra que $d_{vc}(H)<D+2$. 

Luego, de (a) y (b) se conluye que $d_{vc}(H)=D+1$.

### Problema 2.20 

Hay varios límites en la generalización error $\epsilon$, todos con probabilidad de al menos $1-\delta$.

(a) Cota VC original:

$$\epsilon\leq \sqrt{\frac{8}{N}\ln{\frac{4(m_{H}(2N))}{\delta}}}.$$

(b) Cota Penalti de Rademacher:

$$\epsilon\leq \sqrt{\frac{2}{N}\ln{(2N(m_{H}(N))})}+\sqrt{\frac{2}{N}\ln{\frac{1}{\delta}}}+\frac{1}{N}.$$

(c) Parrondo y Van den Broek:

$$\epsilon\leq \sqrt{\frac{1}{N}(2\epsilon+\ln{\frac{6m_h(2N)}{\delta}})}.$$

(d) Devroye:
 
$$\epsilon\leq \sqrt{\frac{1}{2N}(4\epsilon(1+\epsilon)+\ln{\frac{4m_h(N^2)}{\delta}})}.$$

Tenga en cuenta que (c) y (d) son límites implícitos en $\epsilon$. Fije $d_{vc}=50$ y $\delta=0.05$ y grafique estos límites como una función de $N$. ¿Cuál es mejor? 

### Solución

Para utilizar $d_{vc}=50$, en cada cota se reemplaza $m_H(N)$ por su cota superior, $N^{d_{vc}}+1$. Observe que la cota de Deveroye involucra el termino $N^{2d_{vc}}=$N^{100}, entonces con valores de $N$ relativamente pequeños el algoritmo que calcula esta cota se puede hacer un desbordamiento de memoria. Calculemos empiricamente con que valor maximo de $N$ podemos ejecutar los algoritmos:

```python
dvc=50
delta=0.05

#se definen la función de crecimiento y las funciones de las cotas
def mH(N,dcv):
    return N**dcv +1

def VC(N,delta,dvc):
    return np.sqrt(8/N*np.log(4*mH(2*N,dvc)/delta))

def penalti(N,delta,dvc):    
    return np.sqrt(np.log(2*N*mH(N,dvc)/1)*2/N)+np.sqrt(np.log(1/delta)*2/N)+1/N

def parrondo(N,delta,dvc,tolerancia=0.0001,epsi0=0,control=1):
    epsilon=np.sqrt((2*epsi0+np.log(6*mH(2*N,dvc)/delta))*1/N)
    while np.abs(epsilon-epsi0)>tolerancia and control<10000:
                    epsi0=epsilon
                    epsilon=np.sqrt((2*epsi0+np.log(6*mH(2*N,dvc)/delta))*1/N)
                    control+=1    
    return epsilon

def  Devroye(N,delta,dvc,tolerancia=0.0001,epsi0=0,control=1):
    epsilon=np.sqrt(1/(2*N)*(4*epsi0*(1+epsi0)+np.log((4*mH(N**2,dvc)/delta))))
    while np.abs(epsilon-epsi0)>tolerancia and control<10000:
                    epsi0=epsilon
                    epsilon=np.sqrt(1/(2*N)*(4*epsi0*(1+epsi0)+np.log((4*mH(N**2,dvc)/delta))))
                    control+=1    
    return epsilon

# se establecen los datos para graficar
N=1000 # con un N más grande la cota de Devroye hace desbordamiento de memoria
Nvector=list(range(3,N))
epsilonVC,epsilonPenalti,epsilonParrondo,epsilonDevroye=[],[],[],[]

for i in Nvector:
    epsilonVC.append(VC(i,delta,dvc))
    epsilonPenalti.append(penalti(i,delta,dvc))
    epsilonParrondo.append(parrondo(i,delta,dvc))
    epsilonDevroye.append(Devroye(i,delta,dvc))
    

plt.plot(Nvector,epsilonVC,'g',label='$Cota VC$')
plt.plot(Nvector,epsilonPenalti,'b',label='$Cota Penalti$')
plt.plot(Nvector,epsilonParrondo,'r',label='$Cota Parrodo-Van den Broek$')
plt.plot(Nvector,epsilonDevroye,'y',label='$Cota de Devroye$')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.xlabel('$N$')
plt.ylabel('$𝜖$')
plt.axhline(1, color="black", label='$𝜖=1$')
plt.yscale('log')#Se grafica a escala para que se aprecie mejor las curvas
plt.title("COTAS DE ML", 
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 18})
```

Observamos que las cotas de Devroye y Parrondo-Van den Broek decren más rapido por tanto son mejores al momento de estimar los datos necesarios para que, con una certeza de almenos $1-\delta$, tener como máximo un error de generalización $\epsilon$; para $\epsilon$ y $\delta$ dados. En la grafica vemos que con $1000$ datos la cota VC esta lejos de alcanzar $\epsilon=1$, la cota "Penalti" apenas pasa $\epsilon=1$, mientras las cotas de Devroye y Parrondo-Van den Broek ya estan por debajo de $\epsilon=0.5$.

Si queremos graficar con un $N$ más grande, para ver en qué momento la cota VC cruza a $\epsilon=1$, podemos despejar $\epsilon$ de la cota de Devroye y calcular la sin iterar.

Despejemos $\epsilon$ de la cota de Devroye:
\begin{equation}
\begin{split}
\epsilon &\leq \sqrt{\frac{1}{2N}(4\epsilon(1+\epsilon)+\ln{\frac{4m_h(N^2)}{\delta}})},\\
N\epsilon^2 & \leq 2\epsilon^2+2\epsilon +\frac{1}{2}\ln{\frac{4m_h(N^2)}{\delta}},\\
\epsilon^2(N-2)-2\epsilon & \leq \frac{1}{2}\ln{\frac{4m_h(N^2)}{\delta}},\\
\epsilon^2-\frac{2}{(N-2)}\epsilon & \leq \frac{1}{2(N-2)}\ln{\frac{4m_h(N^2)}{\delta}},\\
\epsilon^2-\frac{2}{(N-2)}\epsilon+\frac{1}{(N-2)^2} & \leq \frac{1}{2(N-2)}\ln{\frac{4m_h(N^2)}{\delta}}+\frac{1}{(N-2)^2},\\
(\epsilon-\frac{1}{(N-2)})^2 & \leq \frac{1}{2(N-2)}\ln{\frac{4m_h(N^2)}{\delta}}+\frac{1}{(N-2)^2},\\
\epsilon-\frac{1}{(N-2)} & \leq \sqrt{\frac{1}{2(N-2)}\ln{\frac{4m_h(N^2)}{\delta}}+\frac{1}{(N-2)^2}},\\
\epsilon & \leq \sqrt{\frac{1}{2(N-2)}\ln{\frac{4m_h(N^2)}{\delta}}+\frac{1}{(N-2)^2}}+\frac{1}{(N-2)},\\
\epsilon & \leq \sqrt{\frac{1}{2(N-2)}\ln{\frac{4(N^{2d_{vc}}+1))}{\delta}}+\frac{1}{(N-2)^2}}+\frac{1}{(N-2)}.
\end{split}
\end{equation}

Ahora modifiquemos la función correpondiente y grafiquemos con en $N$ más grande.

```python
dvc=40
def  Devroye2(N,delta,dvc):
    epsilon=np.sqrt(1/(2*(N-2))*np.log((4*mH(N**2,dvc)/delta))+1/(N-2)**2) +1/(N-2)     
    return epsilon
N=7000 # con un N más grande la cota de Devroye hace desbordamiento de memoria
Nvector=list(range(3,N))
epsilonVC,epsilonPenalti,epsilonParrondo,epsilonDevroye2=[],[],[],[]

for i in Nvector:
    epsilonVC.append(VC(i,delta,dvc))
    epsilonPenalti.append(penalti(i,delta,dvc))
    epsilonParrondo.append(parrondo(i,delta,dvc))
    epsilonDevroye2.append(Devroye2(i,delta,dvc))
    

plt.plot(Nvector,epsilonVC,'g',label='$Cota VC$')
plt.plot(Nvector,epsilonPenalti,'b',label='$Cota Penalti$')
plt.plot(Nvector,epsilonParrondo,'r',label='$Cota Parrodo-Van den Broek$')
plt.plot(Nvector,epsilonDevroye2,'y',label='$Cota de Devroye$')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.xlabel('$N$')
plt.ylabel('$𝜖$')
plt.axhline(1, color="black", label='$𝜖=1$')
plt.yscale('log')#Se grafica a escala para que se aprecie mejor las curvas
plt.title("COTAS DE ML", 
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 18})
```

Al despejar $\epsilon$ de la cota de Devroye no se solucionó el problema de desbordamiento de memoria para $N\geq 1100$; el verdadero problema es que $ln{\frac{4(N^{2d_{vc}}+1))}{\delta}}$ se torna demasiado grande con $d_{vc}$ muy grande. Para graficar con $N$ más grande y ver mejor cómo se comportan las cotas, se hizo $d_{vc}=40$.
